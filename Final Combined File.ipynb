# 4. Methods and Plan
To adress whether or not age can predict the amount of time Minecraft is played, I propose the k-nn reegrssion model be used. As seen in figures 1 and 2, the relationship between age and time played is non linear. We see that individiuals in their late teens and early twenties play far more Minecraft, as compared to other ages. This non-linearity makes the k-nn regression model a good fit for this data set as it predicts values based off neighboring points, not linear patterns. It is assumed that both variables will be standardized before the model is applied, and for this particular dataset, theres is adequate sampling to represnt each each age range. Limitations include, the choice of K as it can easily overfit or underfit the data, a decrease in perofrmance if the training data is too large, and inaccurate predictions for ages where there may not be enough neighboring points. k will be chosen via 5 fold v fold cross validation. 

Processing of Data:
- Desired variables will be selected (age and time played)
- missing data (N/A) will be removed
- Data will be split (75% Training, 25%testing)
- Training Data will be standardized
- k-nn model will be tuned
- vfold cross validation using RMPSE will be applied to determine k
- k-nn model with determined k will be applied
