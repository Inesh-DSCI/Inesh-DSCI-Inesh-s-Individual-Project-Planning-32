# 3. Exploratory Data Analysis and Visualization of Data from players.csv file: 

Wrangling done:
- players.csv file was already tidy
- sessions.csv was untidy as two columns each contained two variables (start_time and end_time contained both the specific time and the associated date). These two cells were seperated into 4 (chr) variables: date_of_start_time, start_time, date_of_end_time and end_time. For a new total of 7 columns. 



```r
sessions_tidy<- sessions_untidy|> 
separate(col=start_time, into=c("date_of_start_time", "start_time"), sep=" ", convert=TRUE)|>
separate(col=end_time, into=c("date_of_end_time", "end_time"), sep=" ", convert=TRUE)
sessions_tidy
```

Calculation of quantitative variable averages for players.csv: 

```r
mean_values_rounded<-players|>summarize(average_hours_played=mean(played_hours, na.rm = TRUE), average_age=mean(Age,na.rm = TRUE))|>
mutate(average_hours_played=round(average_hours_played,2))|>
mutate(average_age=round(average_age,2))
mean_values_rounded
```

```r
#Figure 1. 
options(repr.plot.width = 14, repr.plot.height = 14) 

time_played_line <- ggplot(players, aes(x=Age, y=played_hours)) + 
geom_line()+theme(text = element_text(size = 20))+labs(title="Figure 1. Age vs. Time Played", x = "Age", y = "Time Played (hours)")

time_played_line
```

```r
#Figure 2. 
options(repr.plot.width = 14, repr.plot.height = 14) 
time_played_gender_line <- ggplot(players, aes(x=Age, y=played_hours, colour= gender)) + 
geom_line()+theme(text = element_text(size = 20))+labs(title="Figure 2. Age vs. Time Played (contributions by each gender)", x = "Age", y = "Time Played (hours)")

time_played_gender_line 
```

```r
#Figure 3. 

options(repr.plot.width = 14, repr.plot.height = 14) 

gender_vs_hours_played<- ggplot(players, aes(x = gender, y = played_hours)) + 
   geom_bar(stat = "identity")+theme(text = element_text(size = 20)) + 
labs(title= "Figure 3. Time Played by Each Gender", x="Gender", y="Time Played (hours)") 
gender_vs_hours_played
```

```r
#Figure 4. 

options(repr.plot.width = 14, repr.plot.height = 14) 

experience_prop<- ggplot(players, aes(x = gender, y = played_hours, fill = experience)) + 
   geom_bar(stat = "identity", position = "fill")+theme(text = element_text(size = 20)) +  
labs(title= "Figure 4. Experience Level Composition of Each Gender", x="Gender", y="Time Played (hours)",  fill = "Experience level") 
experience_prop
```

```r
#Figure 5.

options(repr.plot.width = 14, repr.plot.height = 14) 


age_vs_time_played_plot<- ggplot(players, aes(x=Age, y=played_hours, colour= subscribe, shape= subscribe)) + 
geom_point()+theme(text = element_text(size = 20))+labs(title="Figure 5. Age vs. Time Played (subscriber status)", x = "Age", y = "Time Played (hours)")

age_vs_time_played_plot
```

Relevant Insight gained through vizualization:
- Relationship between age and time played is not linear, vizualization reveals individuals in their late teens or early twenties are correlated with increased hours played as compared to other ages in the data set. 

# 4. Methods and Plan
To adress whether or not age can predict the amount of time Minecraft is played, I propose the k-nn reegrssion model be used. As seen in figures 1, 2 and 5, the relationship between age and time played is non linear, making the k-nn regression model a good fit for this data set as it predicts values based off neighboring points. It is assumed that both variables will be standardized before the model is applied, and there is adequate sampling to represnt each each age range. Limitations include, a decrease in performance caused by a large amount of training data, and inaccurate predictions for values outside of the data range. 

Processing of Data:
- Desired variables will be selected (age and time played)
- Data will be split (75% Training, 25%testing)
- Data will be standardized
- k-nn model will be tuned
- vfold cross validation using RMPSE will be applied to determine k
- k-nn model with determined k will be applied
