# 2.Question + Results from 3. Exploratory Data Analysis and Visualization
The broad question chosen for this data analysis is question 2: "We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts." To answer this question I will be focussing on determining whether or not the age of the player can predict the amount of hours of Mincecraft played. In theory, if a relationship between age and hours played is established, we can find what age or "kind" or player contributes the most data (plays the most minecraft) and target them. 

```r
sessions_tidy<- sessions_untidy|> 
separate(col=start_time, into=c("date_of_start_time", "start_time"), sep=" ", convert=TRUE)|>
separate(col=end_time, into=c("date_of_end_time", "end_time"), sep=" ", convert=TRUE)
sessions_tidy
```

Wrangling done:
-players.csv file was already tidy as each coloumn contained one variable, each row contained one observation and each cell contained one value
-sessions.csv was untidy as two columns each contained two variables (start_time and end_time contained both the specific time and the associated date), which also means that cells contained more than one value. These two cells were seperated into 4 <chr> variables: date_of_start_time, start_time, date_of_end_time and end_time. For a new total of 7 columns. 



Calculation of quantitative variable averages for players.csv: 

```r
mean_values_rounded<-players|>summarize(average_hours_played=mean(played_hours, na.rm = TRUE), average_age=mean(Age,na.rm = TRUE))|>
mutate(average_hours_played=round(average_hours_played,2))|>
mutate(average_age=round(average_age,2))
mean_values_rounded
```

# 4. Methods and Plan
To adress whether or not age can predict the amount of time Minecraft is played, I propose the k-nn reegrssion model be used. As seen in figures 1 and 2, the relationship between age and time played is non linear. We see that individiuals in their late teens and early twenties play far more Minecraft, as compared to other ages. This non-linearity makes the k-nn regression model a good fit for this data set as it predicts values based off neighboring points, not linear patterns. It is assumed that both variables will be standardized before the model is applied, and for this particular dataset, theres is adequate sampling to represnt each each age range. Limitations include, the choice of K as it can easily overfit or underfit the data, a decrease in perofrmance if the training data is too large, and inaccurate predictions for ages where there may not be enough neighboring points. k will be chosen via 5 fold v fold cross validation. 

Processing of Data:
- Desired variables will be selected (age and time played)
- missing data (N/A) will be removed
- Data will be split (75% Training, 25%testing)
- Training Data will be standardized
- k-nn model will be tuned
- vfold cross validation using RMPSE will be applied to determine k
- k-nn model with determine k will be applined
